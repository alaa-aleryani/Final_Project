```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
# Read the data into a Pandas DataFrame
ecommerce_df = pd.read_excel('../Resources/Sample_Furniture_Ecommerce.xlsx')
print(ecommerce_df.shape)
ecommerce_df.head()
```

```python
# Get the df columns.
ecommerce_df.columns
```

```python
# Drop unwanted columns
ecommerce_df_cleaned = ecommerce_df.drop(columns = ["Row ID", "Order ID", "Customer ID", 
                                                    "Customer Name", "Product ID", "Product Name"])

print(ecommerce_df.shape)             # Before dropping the six columns
print(ecommerce_df_cleaned.shape)     # After dropping the 6 columns.
ecommerce_df_cleaned.head()           ## NOTE: drop city, state, and postal code for ML
```

```python
# Add a "Sales Per Unit" and "Original Price" columns
# Price Per Item
ecommerce_df_cleaned["Sales Per Unit"] = ecommerce_df_cleaned["Sales"] / ecommerce_df_cleaned["Quantity"]

# Original Price
ecommerce_df_cleaned["Original Unit Price"] = ecommerce_df_cleaned["Sales Per Unit"] / (1 - ecommerce_df_cleaned["Discount"])
print(ecommerce_df_cleaned.shape)
ecommerce_df_cleaned.head(5)
```

```python
# Get the unique categories and subcategories in separate lists.
categories = ecommerce_df_cleaned["Category"].unique()
subcategories = ecommerce_df_cleaned["Sub-Category"].unique()

print(f"There're {len(categories)} categories: \n {categories}")
print("------------" * 6)
print(f"There're {len(subcategories)} subcategories: \n {subcategories}")
```

```python
# Get the unique "Ship Mode", "Segment", and "Region" in separate lists.
ship_mode = ecommerce_df_cleaned["Ship Mode"].unique()
segments = ecommerce_df_cleaned["Segment"].unique()
regions = ecommerce_df_cleaned["Region"].unique()


print(f"There're {len(ship_mode)} ship modes: \n {ship_mode}")
print("------------" * 6)
print(f"There're {len(segments)} segments: \n {segments}")
print("------------" * 6)
print(f"There're {len(regions)} segments: \n {regions}")
```
